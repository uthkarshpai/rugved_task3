```python
def compute_error(X,Y,w,b):
    m=X.shape[0]
    cost=0.0
    for i in range(m):
        fwb_i=np.dot(X[i],w)+b
        cost=cost+(fwb_i-Y[i])**2
    cost=cost/(2*m)
    return cost
```

```python
compute_error(X,Y,w,b=0)
```

```python
def compute_derivative(X , Y , w , b):
    m , n = X.shape
    dj_dw = np.zeros(n,)
    dj_db = 0
    for i in range(m):
        error = (np.dot(X[i], w) + b) - Y[i]
        for j in range(n):
            dj_dw[j] = dj_dw[j] + error * X[i, j] 
        dj_db = dj_db + error
    dj_dw = dj_dw/m
    dj_db = dj_db/m
    return dj_dw , dj_db

```

```python
compute_derivative(X,Y,w,b=0)
```

```python
def compute_gradient(X , Y , w , b , compute_error , compute_derivative , alpha , num_iters ):
    cost_history=[]
    for i in range(num_iters):
        dj_dw,dj_db=compute_derivative(X,Y,w,b)
        w=w-alpha*dj_dw
        b=b-alpha*dj_db

        if i<1000:
            cost_history.append(compute_error(X,Y,w,b))
        if i%ma.ceil(num_iters/10)==0:
            print(f"iteration {i:4d} : Cost {cost_history[-1]:8.2f}")
    return w,b,cost_history
    
   
```

```python
w=np.zeros(26)
b=0.0
num_iters=100
alpha=0.001
w_final,b_final,cost_final=compute_gradient(X,Y,w,b,compute_error,compute_derivative,alpha,num_iters)
print(f"b,w found by our function:{b_final:0.2f},{w_final}")
```

```python
m,_=X.shape
for i in range(m):
    print(f"prediction :{np.dot(X[i],w_final)+b_final}, actual values:{Y[i]}")
```

```python
def accuracy(X,Y,w_final,b_final):
    y_dash=np.dot(X,w_final)+b_final
    errornum=0
    errorden=0
    for i in range(len(Y)):
        errornum+=(Y[i]-y_dash[i])**2
        errorden+=(Y[i]-np.mean(Y))**2
    accu=1-(errornum/errorden)

    print(f"model has an accuracy of {accu}")
```

```python
accuracy(X,Y,w_final,b_final)
```
